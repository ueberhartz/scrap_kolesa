```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

url = "https://kolesa.kz/cars/karaganda/"
driver.get(url)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, 'html.parser')

car_cards = soup.find_all("div", class_="a-card__info")
links = []

while True:
   
    time.sleep(3) 
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    car_cards = soup.find_all("div", class_="a-card__info")
    for card in car_cards:
        link_element = card.find("a", class_="a-card__link")
        if link_element:
            link = link_element.get('href')
            if link and link not in links:  
                links.append(link)

    next_button = driver.find_elements(By.LINK_TEXT, "следующая")

    if next_button and next_button[0].is_enabled():
        next_button[0].click()
    else:
        break

print(f"Total unique links: {len(links)}")
links
```

```python
with open("car_links.txt", "w", encoding="utf-8") as file:
    for link in links:
        file.write(f"https://kolesa.kz{link}\n")
```

```python
import random
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time

driver = webdriver.Firefox()

with open("car_links.txt", "r", encoding="utf-8") as file:
    links = [line.strip() for line in file.readlines()]

links = links[200:700] # 200 - 700 (500)

df = pd.DataFrame(columns=['Price', 'Generation', 'Body', 'Engine', 'Transmission', 'Drive', 'Steering Wheel', 'Color', 'Customs'])
rows = []

def get_element_text(driver, by, value):
    try:
        element = driver.find_element(by, value)
        return element.text.strip()
    except NoSuchElementException:
        return None

processed_count = 0

for link in links:
    try:
        driver.get(link)

        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CLASS_NAME, 'offer__price'))
        )

        # Сбор данных
        price = get_element_text(driver, By.CLASS_NAME, 'offer__price')
        generation = get_element_text(driver, By.XPATH, "//dt[@title='Поколение']/following-sibling::dd")
        body = get_element_text(driver, By.XPATH, "//dt[@title='Кузов']/following-sibling::dd")
        engine = get_element_text(driver, By.XPATH, "//dt[@title='Объем двигателя, л']/following-sibling::dd")
        transmission = get_element_text(driver, By.XPATH, "//dt[@title='Коробка передач']/following-sibling::dd")
        drive = get_element_text(driver, By.XPATH, "//dt[@title='Привод']/following-sibling::dd")
        steering_wheel = get_element_text(driver, By.XPATH, "//dt[@title='Руль']/following-sibling::dd")
        color = get_element_text(driver, By.XPATH, "//dt[@title='Цвет']/following-sibling::dd")
        customs = get_element_text(driver, By.XPATH, "//dt[@title='Растаможен в Казахстане']/following-sibling::dd")

        rows.append([price, generation, body, engine, transmission, drive, steering_wheel, color, customs])
        processed_count += 1
        print(f"Processed: {processed_count}/{len(links)}")

        time.sleep(random.uniform(1, 5))

    except Exception as e:
        print(f"Error processing {link}: {e}")
        continue

driver.quit()

df = pd.DataFrame(rows, columns=['Price', 'Generation', 'Body', 'Engine', 'Transmission', 'Drive', 'Steering Wheel', 'Color', 'Customs'])
```
